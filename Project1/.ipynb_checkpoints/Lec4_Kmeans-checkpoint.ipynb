```python
# import data
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
#data = pd.read_csv('Data_1.csv')
data = pd.read_csv('Data_2.csv')
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['x1'], data['x2'], marker='.', color='black', s=20)
plt.xlabel('X1', fontsize=10)
plt.ylabel('X2', fontsize=10)
plt.grid(True, alpha=0.3, linestyle="--")
plt.show()
```

```python
#Specify an inputfor the process of implementation
data_array = np.column_stack([data['x1'], data['x2']])
```

```python
#Implement the K-means algorithm
number_of_cluster = 3
kmeans_algorithm = KMeans(n_clusters=number_of_cluster, random_state=0).fit(data_array)
kmeans_labels = kmeans_algorithm.labels_
kmeans_cluster_cennter = kmeans_algorithm.cluster_centers_
```

```python
#Visualize the results of the K-Means algorithm
plt.scatter(data['x1'], data['x2'], c = kmeans_labels, marker='.', cmap='tab20', s=20)
plt.xlabel('X1', fontsize=10)
plt.ylabel('X2', fontsize=10)
plt.grid(True, alpha=0.3, linestyle='--')
plt.show()

#Kmeans 의 한계
```

# Lec 5

```python
# import data
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
#data = pd.read_csv('Data_3.csv')
data = pd.read_csv('Data_5.csv')
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['x1'], data['x2'], marker='.', color='black', s=20)
plt.xlabel('X1', fontsize=10)
plt.ylabel('X2', fontsize=10)
plt.grid(True, alpha=0.3, linestyle="--")
plt.show()
```

```python
#Hyper-parameters for the DBSCAN algorithm
dbscan_eps = 0.2
dbscan_min_sample = 2

#specify an input for the process of implementation
data_array = np.column_stack([data['x1'], data['x2']])
```

```python
#Implement the DBSCAN algorithm
from sklearn.cluster import DBSCAN
dbscan_algorithm = DBSCAN(eps=dbscan_eps, min_samples=dbscan_min_sample, metric='euclidean', 
                          algorithm='ball_tree').fit(data_array)
dbscan_labels = dbscan_algorithm.labels_
```

```python
#Count number of clusters in labels
dbscan_number_of_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
dbscan_number_of_clusters
```

```python
#Count number of noise points
dbscan_number_of_noise = list(dbscan_labels).count(-1)
dbscan_number_of_noise
```

```python
#VIsualize the results of the DBSCAN algorith
plt.scatter(data['x1'], data['x2'], c=dbscan_labels, marker='.', cmap='tab20', s=20)
plt.xlabel('X1', fontsize=10)
plt.ylabel('X2', fontsize=10)
plt.grid(True, alpha=0.3, linestyle="--")
#plt.savefig("Results_DBSCAN", dpi=250, bbox_inches="tight")
plt.show()
```
